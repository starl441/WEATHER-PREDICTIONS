```python
import pandas as pd
microsoft_df=pd.read_csv('Microsoft_Data.csv')

Total = microsoft_df.isnull().sum().sort_values(ascending=False)
Percentage = (microsoft_df.isnull().sum()*100)/microsoft_df.isnull().count().sort_values(ascending=False)
missing_values = pd.concat([Total,Percentage],keys=['Total','Percentage'],axis=1,sort=False).sort_values('Total',ascending=False)
missing_values
microsoft_df.drop(columns=['MachineIdentifier','PuaMode','Census_ProcessorClass','DefaultBrowsersIdentifier','Census_IsFlightingInternal','Census_InternalBatteryType','Census_ThresholdOptIn','Census_IsWIMBootEnabled'],inplace=True)
        

    
for col in microsoft_df.columns:
    if microsoft_df[col].isnull().sum() > 10000:
        microsoft_df.drop(col,axis=1,inplace=True)


microsoft_df_dtypes={'object':[col for col in microsoft_df.columns if microsoft_df[col].dtype =='object'],'int or float':[col for col in microsoft_df.columns if microsoft_df[col].dtype =='float64']}

for col in microsoft_df_dtypes['object']:
    microsoft_df[col].fillna(value=microsoft_df[col].mode()[0],inplace=True)
    
for col in microsoft_df_dtypes['int or float']:
    microsoft_df[col].fillna(value=microsoft_df[col].median(),inplace=True)
    
```

```python
from sklearn.model_selection import train_test_split
y=microsoft_df['HasDetections']
microsoft_df.drop(columns='HasDetections',inplace=True)


X_train,X_test,y_train,y_test=train_test_split(microsoft_df,y,test_size=0.3,random_state=40)
X_train
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
for col in microsoft_df_dtypes['object']:
    le.fit(X_train[col])
    X_test[col] = X_test[col].map(lambda s: '<unknown>' if s not in le.classes_ else s)
    le.classes_ = np.append(le.classes_, '<unknown>')
    X_train[col] = le.transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
    
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
X_train_scaled= MMS.fit_transform(X_train)
X_test_scaled= MMS.transform(X_test)
X_train_scaled
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca = PCA().fit(X_train_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
#plt.plot((pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
pca.components_
plt.show()
```

```python
column_list= range(0,40)
pca_1 = PCA(n_components=40)
X_train_new = pca_1.fit_transform(X_train_scaled)
df_train = pd.DataFrame(data=X_train_new, columns=column_list)
df_train
```

```python
X_test_new = pca_1.transform(X_test_scaled)
df_test = pd.DataFrame(data=X_test_new,columns=column_list)
df_test
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(df_train,y_train)
y=y.astype('int')
predicted = LR.predict(df_test)
predicted
```

```python
from sklearn.metrics import f1_score,precision_score,confusion_matrix
f1_score(y_test,predicted)
```
